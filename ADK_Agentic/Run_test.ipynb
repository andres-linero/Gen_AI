```python
folder = "/Users/andreslinero/Desktop/Full Time Dev/Gen_AI/Scans/incoming"

# Step 2: Look at all files in the folder
files = os.listdir(folder)
print("Files found:", files)

# Step 3: Get full path for each file
full_paths = []
for file in files:
    full_path = os.path.join(folder, file)
    full_paths.append(full_path)
    
print("Full paths:", full_paths)

# Step 4: Sort by time (when file was modified)
sorted_files = sorted(full_paths, key=lambda f: os.path.getmtime(f))
print("Sorted by time:", sorted_files)

pairs = []
for i in range(0, len(sorted_files), 2):
    if i + 1 < len(sorted_files):
        pair = {
            "ticket": sorted_files[i],
            "id": sorted_files[i + 1]
        }
        pairs.append(pair)

print("Pairs:", pairs)
```
